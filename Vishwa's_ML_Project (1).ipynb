```python
from sklearn.preprocessing import StandardScaler, LabelEncoder

# Encode categorical variables
label_encoder_gender = LabelEncoder()
df['Gender'] = label_encoder_gender.fit_transform(df['Gender'])

label_encoder_marital_status = LabelEncoder()
df['Marital_Status'] = label_encoder_marital_status.fit_transform(df['Marital_Status'])

# Standardize the data
scaler = StandardScaler()
data_scaled = scaler.fit_transform(df)

# Convert the standardized data back to a DataFrame
data_scaled_df = pd.DataFrame(data_scaled, columns=df.columns)

# Display the first few rows of the processed dataset
data_scaled_df.head()
```

```python
print(data_scaled.shape)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Define the number of clusters
num_clusters = 33  # This can be adjusted based on the dataset

# Train the K-means model
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
kmeans.fit(data_scaled_df)

# Predict the cluster for each data point
labels = kmeans.predict(data_scaled_df)

# Evaluate the model
inertia = kmeans.inertia_
silhouette_avg = silhouette_score(data_scaled_df, labels)

print("The inertia report value:", inertia)
print("The Silhouette report value:", silhouette_avg)
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Function to calculate silhouette scores for different number of clusters
def calculate_silhouette_scores(data, cluster_range):
    scores = []
    for n_clusters in cluster_range:
        kmeans = KMeans(n_clusters=n_clusters, random_state=42)
        labels = kmeans.fit_predict(data)
        score = silhouette_score(data, labels)
        scores.append(score)
    return scores

# Define the range for the number of clusters
cluster_range = range(2, 99)

# Calculate silhouette scores for the range of clusters
silhouette_scores = calculate_silhouette_scores(data_scaled_df, cluster_range)

# Plot the silhouette scores
plt.figure(figsize=(10, 6))
plt.plot(cluster_range, silhouette_scores, marker='o')
plt.title('Silhouette Scores for Different Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.grid(True)
plt.show()

# Find the number of clusters with the highest silhouette score
best_num_clusters = cluster_range[silhouette_scores.index(max(silhouette_scores))]
print(f'Best number of clusters: {best_num_clusters}, Silhouette Score: {max(silhouette_scores)}')
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Function to calculate silhouette scores for different numbers of clusters
def calculate_silhouette_scores(data, cluster_range):
    scores = []
    for n_clusters in cluster_range:
        kmeans = KMeans(n_clusters=n_clusters, random_state=42)
        labels = kmeans.fit_predict(data)
        score = silhouette_score(data, labels)
        scores.append(score)
    return scores

# Define the range for the number of clusters
cluster_range = range(2, 99)

# Calculate silhouette scores for the range of clusters
silhouette_scores = calculate_silhouette_scores(data_scaled_df, cluster_range)

# Plot the silhouette scores
plt.figure(figsize=(10, 6))
plt.plot(cluster_range, silhouette_scores, marker='o')
plt.title('Silhouette Scores for Different Number of Clusters (K-Means)')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.grid(True)
plt.show()

# Find the number of clusters with the highest silhouette score
best_num_clusters = cluster_range[silhouette_scores.index(max(silhouette_scores))]
print(f'Best number of clusters: {best_num_clusters}, Silhouette Score: {max(silhouette_scores)}')
```

```python
# Apply K-Means with the best number of clusters
kmeans_optimal = KMeans(n_clusters=best_num_clusters, random_state=42)
labels_kmeans_optimal = kmeans_optimal.fit_predict(data_scaled_df)

# Evaluate the model
silhouette_avg_kmeans_optimal = silhouette_score(data_scaled_df, labels_kmeans_optimal)
print(f'Silhouette Score for K-Means with {best_num_clusters} clusters: {silhouette_avg_kmeans_optimal}')
```

```python
# Split the data into training and testing sets
X_train, X_test = train_test_split(data_subset, test_size=0.3, random_state=42)

# Standardize the data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Determine the optimal number of clusters using the silhouette score
best_num_clusters = 0
best_score = -1
for n_clusters in range(2, 69):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(X_train_scaled)
    score = silhouette_score(X_train_scaled, kmeans.labels_)
    if score > best_score:
        best_num_clusters = n_clusters
        best_score = score

print(f'Optimal number of clusters: {best_num_clusters}')
print(f'Best silhouette score: {best_score}')
```

```python
# Fit the K-Means model with the optimal number of clusters
kmeans_optimal = KMeans(n_clusters=best_num_clusters, random_state=42)
kmeans_optimal.fit(X_train_scaled)
train_labels = kmeans_optimal.labels_

# Predict cluster labels for the test set
test_labels = kmeans_optimal.predict(X_test_scaled)
```

```python
# Calculate silhouette scores for the training and testing sets
silhouette_train = silhouette_score(X_train_scaled, train_labels)
silhouette_test = silhouette_score(X_test_scaled, test_labels)

print(f'Silhouette Score for Training Set: {silhouette_train}')
print(f'Silhouette Score for Testing Set: {silhouette_test}')
```
